# Normalization

```python
from sklearn.preprocessing import MinMaxScaler  #Transform features by scaling each feature to a given range.
```

```python
m= MinMaxScaler()   #object create
```

```python
ms=m.fit(data[['Marketing Spend']])   # fit the data
ms
```

```python
data['Marketing Spend']=m.fit_transform(data[['Marketing Spend']]) #This method performs fit and transform on the input data at a single time and converts the data points.
```

```python
data.head()
```

```python
data1=data.drop(['Area','Profit'],axis=1)
data1.head()
```

```python
data2=data.drop(['Marketing Spend','Administration','Transport'],axis=1)
data2.head()
```

# Using for Loop

```python
for key,value in data1.iteritems():
    ms = m.fit(data1[[key]])
    data1[key] = ms.transform(data1[[key]])
data1.head()
```

# Standardization
